```python
img = cv2.imread(r"C:\Users\hp\Downloads\WhatsApp Image 2026-05-29 at 10.55.31 AM.jpeg")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
text = pytesseract.image_to_string(gray)

print("===== OCR TEXT =====")
print(text)
```

```python
pan = re.findall(r'[A-Z]{5}[0-9]{4}[A-Z]', text)
pan
```
